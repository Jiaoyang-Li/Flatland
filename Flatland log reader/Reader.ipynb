```python
import os
import pandas as pd
```

```python
dataset = []

with open("./Logs/85760-aicrowd-subcontractor.log", 'r') as f:
    lines = f.readlines()
    for l in lines:
        
        if "Evaluating" in l:
#             print(l[l.find("Test_"):l.find("pkl")-1])
            case_name = l[l.find("Test_"):l.find("pkl")-1]
            if(old != case_name):
                print(case_name)
                old = case_name
                
        if "Evaluation finished" in l:
#             print(l)
            # print(int(l[l.find("Evaluation finished in") + len("Evaluation finished in "): l.find(" timesteps")]))
            
            used_time_steps = int(l[l.find("Evaluation finished in") + len("Evaluation finished in "): l.find(" timesteps")])
            
            # print(l[l.find("timesteps, ")+len("timesteps, "): l.find(" seconds.")])
            
            runtime = float(l[l.find("timesteps, ")+len("timesteps, "): l.find(" seconds.")])
            
            # print(l[l.find("Percentage agents done: ")+len("Percentage agents done: "): l.find(". Normalize")])
            
            done_percentage = float(l[l.find("Percentage agents done: ")+len("Percentage agents done: "): l.find(". Normalize")])
            
            # print(l[l.find("Normalized reward: ")+len("Normalized reward: "): l.find(". Number of malfunctions:")])
            
            reward = float(l[l.find("Normalized reward: ")+len("Normalized reward: "): l.find(". Number of malfunctions:")])
                        
            num_malfunc = l[l.find("Number of malfunctions: ")+len("Number of malfunctions: "): ]
            num_malfunc = int(num_malfunc[:num_malfunc.find(".")])

#         print([case_name, num_malfunc, used_time_steps, runtime, done_percentage, reward])
            dataset.append([case_name, num_malfunc, used_time_steps, runtime, done_percentage, reward])
                
dataset
            
        
```

```python
df = pd.DataFrame(dataset, columns = ['case', 'num_malfunction','time_steps','runtime','done_percentage','reward'])
```
